```python
import numpy as np

# Creating arrays
study_hours = np.array([1, 2, 3, 4, 5])
scores = np.array([30, 40, 50, 60, 70])

print("Study Hours:", study_hours)
print("Scores:", scores)

# Indexing & slicing
print("First 3 hours:", study_hours[:3])
print("Last score:", scores[-1])

# Basic math operations
print("Mean study hours:", np.mean(study_hours))
print("Mean score:", np.mean(scores))
print("Standard deviation of scores:", np.std(scores))
print("Max score:", np.max(scores))
print("Min score:", np.min(scores))

# Array-based calculations (vectorized, no loops needed)
predicted_scores = study_hours * 10 + 20
print("Vectorized prediction:", predicted_scores)

# Reshaping (useful later for sklearn, which expects 2D input)
study_hours_2d = study_hours.reshape(-1, 1)
print("Reshaped for ML model:", study_hours_2d)
print("Shape:", study_hours_2d.shape)
```
